```python
class ANOVA:

    def __init__(self, m):
        self.m = m
        self.N = m.size
        self.total_sum = np.sum(m)
        self.sum_sq = np.sum(m**2)

    def SS_total(self):
        return self.sum_sq - (self.total_sum**2 / self.N)

    def SS_between(self):
        group_sums = np.sum(self.m, axis=1)
        n = self.m.shape[1]   # samples per group
        return np.sum(group_sums**2 / n) - (self.total_sum**2 / self.N)

    def SS_within(self):
        return self.SS_total() - self.SS_between()

ANOVA(m)
```
